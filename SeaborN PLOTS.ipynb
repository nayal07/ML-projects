```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
tips.columns
```

```python
tips.info()
```

# 1 SCATTER PLot
A scatter plot is a two-dimensional plot that uses dots to represent the values for two different
variables. It is useful for visualizing the relationship between two continuous variables. Seabornâ€™s
scatterplot() function can be used to create scatter plots, with optional customization of colors,
size, and shape of the markers


```python
plt.figure(figsize=(7,5))

sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Scatter plot")
plt.show()
```

# 2 Line PLot

A line plot is a graph that displays data points connected by straight lines. It is useful for showing trends over time or comparing changes in different groups. Seabornâ€™s lineplot() function can be used to create line
plots, with optional customization of colors, markers, and line styles.


```python
plt.figure.figsize=(5,7)

sns.lineplot(data=tips, x='day', y='total_bill', hue='sex' )
plt.title('Line plot')
plt.show()
```

# 3 Histogram
A histogram is a graphical representation of the distribution of a dataset. It divides the range of values into a set of intervals, and then counts the number of values that fall into each interval.
Seabornâ€™s histplot() function can be used to create histograms, with optional customization of bin size, color, and normalization


```python
plt.figure.figsize=(6,7)

sns.histplot(data=tips, x='total_bill', bins=20)

plt.title("Histogram plot")
plt.show()
```

# 4 BAR PLOT:

A bar plot is a graph that displays categorical data with rectangular bars. It is useful for comparing the values of different categories. Seabornâ€™s barplot() function can be used to create bar plots, with optional customization of colors, error bars, and confidence intervals.

```python
plt.figure.figsize=(6,7)
sns.barplot(data=tips, x='day', y='total_bill', hue='sex')

plt.title("BAR PLOT")
plt.show()
```

# 5 BOX PLOT
A box plot is a graph that displays the distribution of a dataset using five summary statistics: minimum, first quartile, median, third quartile, and maximum. It is useful for identifying outliers and comparing the distributions of different groups. Seabornâ€™s boxplot() function can be used to create box plots, with optional customization of colors, notch, and whisker length.


```python
plt.figure.figsize=(6,7)

sns.boxplot(data=tips , x='day', y='total_bill', hue='sex')
plt.title("BOXPLOT")
plt.show()
```

# 6 HEATMAP

A box plot is a graph that displays the distribution of a dataset using five summary statistics: minimum, first quartile, median, third quartile, and maximum. It is useful for identifying outliers and comparing the distributions of different groups. Seabornâ€™s boxplot() function can be used to
create box plots, with optional customization of colors, notch, and whisker length.


```python
plt.figure.figsize=(6,7)

sns.heatmap(data=iris.corr(), annot=True)
plt.title("HEATMAP")
plt.show()
```

# 7 PAIRPLOT
A pair plot is a graph that displays pairwise relationships between variables in a dataset. It is
useful for exploring the correlations between different variables. Seabornâ€™s pairplot() function can
be used to create pair plots, with optional customization of diagonal plots, colors, and markers.


```python
plt.figure.figsize=(6,7)

sns.pairplot(data= iris, hue='species')
plt.title("PAIRPLOT")
plt.show()
```

# VIOLIN plot
A violin plot is a combination of a box plot and a kernel density plot. It displays the distribution
of a dataset with a combination of a box plot and a density plot, where the density plot shows the
estimated probability density function. Seabornâ€™s violinplot() function can be used to create violin
plots, with optional customization of colors, width, and inner quartile range.

```python
plt.figure(figsize=(6,7))

sns.violinplot(data=tips, x="day", y="total_bill")
plt.title('violin plot')
plt.show()
```

# SwARM PLOT

A swarm plot is a graph that displays the distribution of a dataset by placing each data point
on a line. It is useful for showing the density of data points and highlighting potential outliers.
Seabornâ€™s swarmplot() function can be used to create swarm plots, with optional customization of
colors and size.

```python
plt.figure(figsize=(6,7))

sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')

plt.title("JOINTPLOT")
plt.show()
```

# JOINT PLOT
A joint plot is a graph that displays the relationship between two variables using both a scatter
plot and a histogram. It is useful for exploring the correlations between different variables and
identifying patterns in the data. Seabornâ€™s jointplot() function can be used to create joint plots,
with optional customization of colors, markers, and regression lines.


```python
plt.figure(figsize=(6,7))

sns.jointplot(data=tips , x='total_bill', y='tip', kind='reg')

plt.title("JOINT PLOT")
plt.show()
```
